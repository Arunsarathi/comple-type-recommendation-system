```python
import numpy as np
import pandas as pd
```

```python
col =  ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data',
                 sep='\t',names=col)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python
# we can combain togeter
```

```python
df = pd.merge(df,movie_titles,on='item_id')

# item id is is coman and uniq item so we used
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
rating.sort_values('num of ratings',ascending=False).head(10)
```

```python
rating.head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
terminator_user_ratings = moviemat['Terminator 2: Judgment Day (1991)']
print(starwars_user_ratings.head())
print("................................")
terminator_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_terminator = moviemat.corrwith(terminator_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head()
```

```python
corr_starwars = corr_starwars.join(rating['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_terminator,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(rating['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
user_ratings = moviemat['Terminator 2: Judgment Day (1991)']
similar_to= moviemat.corrwith(user_ratings)
corr_m = pd.DataFrame(similar_to,columns=['Correlation'])
corr_m.dropna(inplace=True)
corr_m = corr_m.join(rating['num of ratings'])
corr_m[corr_m['num of ratings']>100].sort_values('Correlation',ascending=False).head(10)
```
